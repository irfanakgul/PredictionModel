```python
import pandas as pd, numpy as np
import glob
import openpyxl

def fn_concat_trainSet():
    
    # give old pickle data set path and change config file
    old_data_path = './raw202301_41758.pickle'

    # Excel dosyalarını bul
    excel_files = glob.glob('./concat_tables/*.xlsx')  # Klasör yolu ve dosya uzantısını değiştirin

    # Excel dosyalarını DataFrame'lere dönüştür ve birleştir
    df_list = [pd.read_excel(file,engine='openpyxl') for file in excel_files]
    combined_df = pd.concat(df_list, ignore_index=True)

    # Pickle dosyasını yükle
    pickle_df = pd.read_pickle(old_data_path)  # Pickle dosyasının yolu
    print(f'Before concat Dataset size = {len(pickle_df)}')
    # Pickle DataFrame ile birleştir
    final_df = pd.concat([pickle_df, combined_df], ignore_index=True)

    # Sonucu pickle olarak kaydet
    max_date = final_df
    final_df.to_pickle(f'../inputs/train_dataset/raw_{len(final_df)}.pickle')
    print(f'After concat Dataset size = {len(final_df)}\n * DONE *')
    return final_df
```

```python
df = fn_concat_trainSet()
```
